1. Display a list of 5 makes that had the highest number of crashes?

Most personal vehicles involved in crashes were traveling towards which direction?

Answer: North

```python
%%sql
select COUNT(TRAVEL_DIRECTION), TRAVEL_DIRECTION from traffic
where VEHICLE_USE = 'PERSONAL'
GROUP BY TRAVEL_DIRECTION;
```

 How many passengers were involved in personal cars were in accidents?

Answer: 1095

```python
%%sql
select SUM(NUM_PASSENGERS) from traffic
where VEHICLE_USE = 'PERSONAL';
```

How many passengers were involved in accidents while traveling in personal Sport
Utility vehicles?

Answer: None

```python
%%sql
SELECT SUM(NUM_PASSENGERS) FROM traffic
WHERE VEHICLE_USE = 'PERSONAL'
AND VEHICLE_TYPE = 'SPORT UTILITY VEHICLE (SUV) ';
```

Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked.

Answer: 17

```python
%%sql

SELECT COUNT(*) FROM traffic
WHERE MAKE = 'MERCEDES-BENZ'
AND MANEUVER = 'PARKED';
```

 Compute the total occupant count per vehicle count.

Answer: 

```python
%%sql 
SELECT SUM(OCCUPANT_CNT), MAKE FROM traffic
GROUP BY MAKE;
```

Sort the number of passengers in descending order by travel direction taking into account travel direction. Return travel direction and no. of passengers

```python
%%sql

select SUM(NUM_PASSENGERS), TRAVEL_DIRECTION from traffic
GROUP BY TRAVEL_DIRECTION
ORDER BY SUM(NUM_PASSENGERS) DESC;
```

Select the top 5 vehicle models involved in the accident with no defect, order them by make, and show their maneuver.


```python
%%sql

SELECT COUNT(MAKE), MAKE, MANEUVER FROM traffic
WHERE VEHICLE_DEFECT = 'NONE'
GROUP BY MAKE
ORDER BY COUNT(MAKE) DESC
LIMIT 5;
```

Which ridesharing vehicles in an accident had the most number of passengers?
Return model, and the number of passengers.


```python
%%sql
SELECT MODEL,SUM(NUM_PASSENGERS) FROM traffic
where VEHICLE_TYPE = 'PASSENGER'
ORDER BY MAKE;
```
